```python
import pandas as pd

# Load the data
file_path = 'IntGenbyFuel2017.xlsx'
data = pd.read_excel(file_path, sheet_name='Dec')

# Drop unnecessary columns
data = data.drop(data.columns[[0, 2, 3]], axis=1)

# Transpose the DataFrame
transposed_df = data.T.reset_index(drop=True)

# Extract fuel names from the first row
fuel_names = transposed_df.iloc[0]

# Drop the first row from the transposed dataframe
transposed_df = transposed_df.iloc[1:]

# Initialize an empty DataFrame to store time series data
time_series_df = pd.DataFrame()

# Iterate over each unique fuel name
for fuel in fuel_names.unique():
    # Select all columns corresponding to the current fuel
    fuel_columns = transposed_df.loc[:, fuel_names == fuel]

    # Concatenate the columns vertically and reset the index
    concatenated_column = pd.concat([fuel_columns[col] for col in fuel_columns], ignore_index=True)

    # Add this concatenated column to the time series DataFrame
    time_series_df[fuel] = concatenated_column

# Display the first few rows of the final DataFrame
print(time_series_df.head())
```

```python
file_path = 'IntGenByFuel2012.xls'
data = pd.read_excel(file_path, sheet_name='Dec12')

# Create a new column 'Fuel' with only the source name (remove the date)
data['Fuel'] = data.iloc[:, 0].astype(str).str.split('_').str[1]

# Drop the 'date-source' column (assuming it's the first column) and any other unnecessary column
data = data.drop(data.columns[[0, 1]], axis=1)

# Move 'Fuel' column to the first position
data = data[['Fuel'] + [col for col in data.columns if col != 'Fuel']]

# Transpose the DataFrame
transposed_df = data.T.reset_index(drop=True)

# Extract fuel names from the first row
fuel_names = transposed_df.iloc[0]

# Drop the first row from the transposed dataframe
transposed_df = transposed_df.iloc[1:]

# Initialize an empty DataFrame to store time series data
time_series_df = pd.DataFrame()

# Iterate over each unique fuel name
for fuel in fuel_names.unique():
    # Select all columns corresponding to the current fuel
    fuel_columns = transposed_df.loc[:, fuel_names == fuel]

    # Concatenate the columns vertically and reset the index
    concatenated_column = pd.concat([fuel_columns[col] for col in fuel_columns], ignore_index=True)

    # Add this concatenated column to the time series DataFrame
    time_series_df[fuel] = concatenated_column

# Display the first few rows of the final DataFrame
print(time_series_df.head())
```
